```python
import pandas as pd
import re
from fractions import Fraction
import unicodedata
from decimal import Decimal
```

```python
# Load raw data and do some preprocessing
input_data = pd.read_csv("../data/raw/nyt-ingredients-snapshot-2015.csv")
input_data.head()
```

```python
input_data.drop(["index"], axis=1, inplace=True)
input_data.head()
```

```python
# There are leftover html tags that need to be removed
def cleanHTML(s):
    
        m1 = re.findall(r"\s*\(<.*>\)", s) # this will remove all <a href=...>...</a>
        if m1:
            for m in m1:
                #print(s)
                s = re.sub(r"\s*\(<.*>\)", "", s)
                #print(s)
        m2 = re.findall(r"<.*?>", s) # This will remove all <span>
        if m2:
            for m in m2:
                #print(s)
                s = re.sub(r"<.*?>", "", s)
                #print(s)
    except TypeError:
        print(s)
            
input_data["input"] = input_data["input"].apply(cleanHTML)
```

```python
print(input_data.shape)
# There are 144 lines missing an input, these are worthless and are dropped.
input_data.dropna(axis=0, subset=["input"], inplace=True)
print(input_data.shape)
```

```python
def cleanUnicodeFractions(s):
    """
    Replace unicode fractions with ascii representation, preceded by a
    space.

    "1\x215e" => "1 7/8"
    """
    m1 = re.findall(r"(\d+)\s+([\u2150-\u215E\u00BC-\u00BE])", s)
    if m1:
        for m in m1:
            print(s)
            num = float(m[0]) + float(Fraction(unicodedata.numeric(m[1])))
            s = re.sub(r"(\d+)\s+([\u2150-\u215E\u00BC-\u00BE])", str(round(num, 3)), s)
            print(s)
    match = re.findall(r"([\u2150-\u215E\u00BC-\u00BE])", s)
    if match:
        
        for m in match[0]:
            s = re.sub(r""+re.escape(m), str(round(float(Fraction(unicodedata.numeric(m))), 3)), s)
        
    return s

input_data["input"] = input_data["input"].apply(cleanUnicodeFractions)
```

```python
def mergeFractions(sent):
    print(sent)
    m1 = re.findall(r'(\d+)\s+(\d\/\d)', sent)
    if m1 is not None:
        print(sent)
        num = float(m1[0][0]) + float(Fraction(m1[0][1]))
        sent = re.sub(r'(\d+)\s+(\d\/\d)', str(round(num, 3)), sent)
    m2 = re.findall(r'(\d\/\d)', sent)
    if m2 is not None:
        num = float(Fraction(m2[0][0]))
        sent = re.sub(r'(\d\/\d)', str(round(num, 3)), sent)
    return sent

input_data["input"] = input_data["input"].apply(mergeFractions)
```
